```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
file_path = '/content/drive/MyDrive/Data min/activity.csv'
df = pd.read_csv(file_path)
df
```

```python
df['date'] = pd.to_datetime(df['date'],format = '%Y-%m-%d')
```

```python
df.sort_values(by='date', inplace = True)
```

```python
df1 = df
```

# 

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
df1['activity_type']= label_encoder.fit_transform(df1['activity_type'])
```

```python
df1['customer']= label_encoder.fit_transform(df1['customer'])
```

```python
df1['year'] = df1['date'].dt.year
df1['month'] = df1['date'].dt.month
df1['day'] = df1['date'].dt.day

df1.drop('date',axis=1,inplace=True)
```

```python
df1
```

```python
#from sklearn.preprocessing import MinMaxScaler
#mms=MinMaxScaler()
```

```python
x = df1[["customer",'year','month','day']]
x
```

```python
y = df1[['activity_type']]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

# Build a model (ANN) in tensorflow/keras

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(4, input_shape=(4,), activation='relu'),
    #keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=150)
```
